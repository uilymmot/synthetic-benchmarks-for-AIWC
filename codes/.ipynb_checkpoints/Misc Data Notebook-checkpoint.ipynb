```python
os.chdir("/workspace/codes/predictions/Data-Generated_Kernels/")

files = ['gen_radeonvii.csv','gen_gtx2080ti.csv','gen_irispro580.csv']
filesold = ['gen_3900x.csv','gen_gold6134.csv','gen_p100.csv']
df_features = pd.read_csv("gen_aiwcfeatures_abff4149.csv")
df_features['kernel'] = df_features['kernel'].apply((lambda x: x.split("/")[-1]))

data_df = pd.DataFrame()

print("reading in generated files from:", files, filesold)
for f in files:
    ff = pd.read_csv(f)
    ff['kernel_time'] = ff['time']
    ff['total_time'] = ff['time']
    ff['kernel'] = ff['kernel'].apply((lambda x: x.split("/")[-1]))
#     ff['kernel'] = ff['region']
    ff['device'] = f.split("_")[1].split(".")[0]
    ff = ff.drop(['time','region','id'], axis=1)
    
    data_df = pd.concat([data_df, ff])
    
data_df1 = pd.DataFrame()

for f in filesold:
    ff = pd.read_csv(f)
    ff['device'] = f.split("_")[1].split(".")[0]
    data_df1 = pd.concat([data_df1, ff])
    

data_df1 = data_df1[data_df.columns]
data_df = pd.concat([data_df1, data_df])
data_df = data_df.groupby(['kernel','size','device']).head(50).reset_index()

data_df.loc[data_df['device'] == 'gold6134', 'device'] = 'Gold6134'
data_df.loc[data_df['device'] == 'p100', 'device'] = 'P100'
data_df.loc[data_df['device'] == '3900x', 'device'] = '3900x'
data_df.loc[data_df['device'] == 'gtx2080ti', 'device'] = '2080ti'
data_df.loc[data_df['device'] == 'radeonvii', 'device'] = 'RadeonVII'
data_df.loc[data_df['device'] == 'irispro580', 'device'] = 'IrisPro580'

print(np.unique(data_df['device']))

final_df = pd.merge(df_features, data_df, on=['kernel','size'])

print(np.unique(final_df['device']))
final_df['Opcode'] = 0

final_df = final_df.drop(['Unnamed: 0', 'application','index'], axis=1)

# final_df['Mean PSL Gradient'] = np.mean(gradients, axis=1)
# final_df['Max PSL Gradient'] = np.max(gradients, axis=1)
# final_df['Min PSL Gradient'] = np.min(gradients, axis=1)

# final_df = final_df[~(final_df['device'] == '2080ti')]
# final_df = final_df[~(final_df['device'] == 'RadeonVII')]
# final_df = final_df[~(final_df['device'] == 'IrisPro580')]

# final_df = final_df[~(final_df['size'] == 'medium')]


final_df = final_df.groupby(['kernel','size','device']).mean().reset_index()
print("saved file to: gen_compiled_abff4149_single.csv")
final_df.to_csv("gen_compiled_abff4149_singles.csv", index=False)
```
